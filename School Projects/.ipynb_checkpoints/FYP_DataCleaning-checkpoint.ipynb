## Missing and Non-positive Sales and Assets filtering
Standard in corporate finance literature

```python
raw_df.loc[:,['at','sale','ppent','ppegt']].describe()
```

exclude firms with missing or non-positive book value of assets or sales and firms (with less than $5 million in physical capital), as is standard in the literature \
--> physical capital condition drops too many firms, omit this condition

Any firms with missing or negative total assets or sales are included in the drop list

```python
### Size filter applied to RAW Period
all_firms = list(raw_df['gvkey'].unique())
drop_list = []

for firm in tqdm(all_firms):
    # for each firm:
    each = raw_df[raw_df['gvkey'] == firm]
    # check size conditions for the data computation period 1975-2021
    each = each[(1975 <= each['fyear']) & (each['fyear'] <= 2021)] 
    missing_at = each['at'].isnull().sum() # number of missing asset
    missing_sale = each['sale'].isnull().sum() # number of missing sales
    neg_at = (each['at'] <= 0).sum() # number of non-positive asset
    neg_sale = (each['sale'] <= 0).sum() # number of non-positive sales
    low_ppegt = (each['ppegt'] < 5).sum() # number of low ppegt (physical asset defined by method paper)
    
    # Dropping conditions
    if ((missing_at > 0) or (neg_at > 0) or (missing_sale > 0) or (neg_sale > 0)): #and (low_ppegt > 0):
        drop_list.append(firm)

print('firms to be dropped:', len(drop_list))
raw_set = set(raw_df[~raw_df['gvkey'].isin(drop_list)]['gvkey'])
print('surviving:', len(raw_set))

# minimum physical asset condition not used, as it risks losing too many observations
```

```python
### Apply size filter
size_filter = raw_df['gvkey'].isin(drop_list)
dropped = raw_df[size_filter]['gvkey'].nunique()

raw_df = raw_df[~size_filter]
print('Dropped firms:', dropped, f'({round(dropped/len_0 * 100)}%)')
print(raw_df['gvkey'].nunique(), 'firms')

# save temporarily
raw_df.to_parquet('Generated Data/atsale_only.parquet')
raw_df['gvkey'].nunique()
```

# Adjusting Compustat SG&A

**Compustat XSGA = actual SG&A + XRD** \
**In-Process R&D (RDIP)**: externally purchased R&D on products not yet being sold, expensed in Income statement but not added in XSGA (it must be added to SG&A). RDIP appears as negative number, add back to SG&A = subtract negative RDIP \
**Thus for non-missing XSGA, actual SG&A = XSGA - XRD - RDIP** \
**Exception**: Sometimes XRD is allocated to COGS instead of SG&A, detected if XSGA < XRD < COGS: just use XSGA as is for SG&A


We start in 1977 to give firms two years to comply with FASBâ€™s 1975
R&D reporting requirement. If we see a firm with R&D equal to zero or
missing in 1977, we assume the firm was typically not an R&D spender
before 1977, so we set any missing R&D values before 1977 to zero. Otherwise,
before 1977, we either interpolate between the most recent nonmissing
R&D values (if such observations exist) or we use the method in
Appendix A (if those observations do not exist). Starting in 1977, we make
exceptions in cases in which the firmâ€™s assets are also missing. These are
likely years when the firm was privately owned. In such cases, we inter-
polate R&D values using the nearest non-missing values.

```python
### Set gvkey as index to operate calculation on sga of each firms
reg_df.index = reg_df['gvkey']
reg_df = reg_df.drop(columns=['gvkey'])

raw_df.index = raw_df['gvkey']
raw_df = raw_df.drop(columns=['gvkey'])
```

```python
### REG DATA
# Rectify non-R&D SG&A - REG DATA
# if xsga < xrd < cogs, xrd is in cogs, use xsga for SG&A
original_sga = (reg_df['xrd'] > reg_df['xsga']) & (reg_df['xrd'] < reg_df['cogs'])

# compute adjustment using non-original portion of sga
non_original_sga = original_sga.apply(lambda x: not x)
# Those not using original xsga must adjust xsga for xrd and rdip - REG DATA
reg_df.loc[non_original_sga,'xsga'] = reg_df.loc[non_original_sga,'xsga'] \
                                    - reg_df.loc[non_original_sga,'xrd']  \
                                    - reg_df.loc[non_original_sga,'rdip']
```

```python
### RAW DATA
# Rectify non-R&D SG&A - RAW DATA
# if xsga < xrd < cogs, xrd is in cogs, use xsga for SG&A
original_sga = (raw_df['xrd'] > raw_df['xsga']) & (raw_df['xrd'] < raw_df['cogs'])

# compute adjustment using non-original portion of sga
non_original_sga = original_sga.apply(lambda x: not x)
# Those not using original xsga must adjust xsga for xrd and rdip - RAW DATA
raw_df.loc[non_original_sga,'xsga'] = raw_df.loc[non_original_sga,'xsga'] \
                                    - raw_df.loc[non_original_sga,'xrd']  \
                                    - raw_df.loc[non_original_sga,'rdip']
```

```python
### reset index, remove gvkey index
reg_df = reg_df.reset_index()
raw_df = raw_df.reset_index()
```

# Create robust_raw dataset for survivorship-robust analysis
- reg_df: firms for regression from 2009-2021 - subject to significant survivorship bias
- raw_df: firms (same firms as reg_df) for intangibles from 1975-2021 - subject to significant survivorship bias
- robust_raw: firms (more firms than reg_df and raw_df) for survivorship-robust regression from 1975-2021 - subject to mild survivorship bias

```python
# Check Number of firms in Reg and Raw
reg_df['gvkey'].nunique(), raw_df['gvkey'].nunique()
reg_df_keys = list(reg_df['gvkey'].unique())
raw_df_keys = list(raw_df['gvkey'].unique())

# symmetric difference of sets = union - intersection
union = list(set(reg_df_keys) | set(raw_df_keys))
intersection = list(set(reg_df_keys) & set(raw_df_keys))
sym_difference = list(set(reg_df_keys) ^ set(raw_df_keys))
only_in_reg = list(set(reg_df_keys) - set(raw_df_keys))
only_in_raw = list(set(raw_df_keys) - set(reg_df_keys))

print(len(reg_df_keys), len(raw_df_keys))
print('survived sample in reg =', len(intersection)) 
print('removed from total sample:',
      len(sym_difference), 
      len(set(union) - set(intersection)),
      len(set(only_in_reg) | set(only_in_raw)))

# mismatch of 2 are out of 3 dropped from Raw due to missing cogs
print('Abnormal data in reg but not in raw:',len(only_in_reg)) # equals 0 if reg is complete subset of raw
print('Lost from raw subject to survivorship bias:',
      len(only_in_raw),
      f'({round(len(only_in_raw)/len(raw_df_keys) * 100, 2)}%)')

### SET ASIDE RAW DATA FOR ROBUSTNESS TEST
robust_raw = raw_df.copy()
```

```python
# match tickers of RAW DATA with REG DATA
intersection_raw = raw_df['gvkey'].isin(intersection)
intersection_reg = reg_df['gvkey'].isin(intersection)

raw_df = raw_df[intersection_raw]
reg_df = reg_df[intersection_reg]
```

```python
### Raw Data same as Reg data tickers
print(robust_raw['gvkey'].nunique(), 'firms') # used for survivorship bias robustness testing
print(raw_df['gvkey'].nunique(), 'firms') # used for intangibles variable calculation
print(reg_df['gvkey'].nunique(), 'firms') # used for regression
print(raw_df.shape, reg_df.shape)
print('Reg periods:', reg_df['fyear'].nunique())
print('Raw periods:', raw_df['fyear'].nunique())
```

# Computing Knowledge and Organization capital
The BEAâ€™s R&D depreciation rates are from the analysis of Li (2012).
The depreciation rates range from 10% in the pharmaceutical industry to
40% for computers and peripheral equipment. Following the BEAâ€™s guidance,
we use a depreciation rate of 15% for industries not in Liâ€™s.
Our results are virtually unchanged if we set Î´R&D equal to 10%, 15%, or
20% for all industries.

```python
#### BEA industry for R&D depreciation rates
### REG
for industry in ffi.BEA_SIC:
    for code_range in ffi.BEA_SIC[industry]:
        if type(code_range) == tuple:
            lower = int(code_range[0])
            upper = int(code_range[1])
            #print('range',lower, upper)
            reg_df['sic_code'] = reg_df['sic_code'].apply(lambda x: ffi.broadcast(x, lower, upper, industry))
        elif type(code_range) == int:
            #print('single',code_range)
            reg_df['sic_code'] = reg_df['sic_code'].apply(lambda x: industry if x == code_range else x)

reg_df['sic_code'] = reg_df['sic_code'].apply(lambda x: 'Other' if type(x) == int else x)
reg_df['sic_code'].unique()
print('Reg----------')
print(reg_df['sic_code'].value_counts()/len(reg_df['sic_code']) * 100)

### RAW
for industry in ffi.BEA_SIC:
    for code_range in ffi.BEA_SIC[industry]:
        if type(code_range) == tuple:
            lower = int(code_range[0])
            upper = int(code_range[1])
            #print('range',lower, upper)
            raw_df['sic_code'] = raw_df['sic_code'].apply(lambda x: ffi.broadcast(x, lower, upper, industry))
        elif type(code_range) == int:
            #print('single',code_range)
            raw_df['sic_code'] = raw_df['sic_code'].apply(lambda x: industry if x == code_range else x)

raw_df['sic_code'] = raw_df['sic_code'].apply(lambda x: 'Other' if type(x) == int else x)
raw_df['sic_code'].unique()
print('\nRaw----------')
print(raw_df['sic_code'].value_counts()/len(raw_df['sic_code']) * 100)

### ROBUST_RAW
for industry in ffi.BEA_SIC:
    for code_range in ffi.BEA_SIC[industry]:
        if type(code_range) == tuple:
            lower = int(code_range[0])
            upper = int(code_range[1])
            #print('range',lower, upper)
            robust_raw['sic_code'] = robust_raw['sic_code'].apply(lambda x: ffi.broadcast(x, lower, upper, industry))
        elif type(code_range) == int:
            #print('single',code_range)
            robust_raw['sic_code'] = robust_raw['sic_code'].apply(lambda x: industry if x == code_range else x)

robust_raw['sic_code'] = robust_raw['sic_code'].apply(lambda x: 'Other' if type(x) == int else x)
robust_raw['sic_code'].unique()
print('\nRobust_Raw----------')
print(robust_raw['sic_code'].value_counts()/len(robust_raw['sic_code']) * 100)
```

```python
### Setting R&D Depreciation rates
## REG
reg_df['rnd_dep'] = None
for industry in ffi.BEA_rnd_dep:
    reg_df.loc[reg_df['sic_code'] == industry,'rnd_dep'] = ffi.BEA_rnd_dep[industry]
# Rest of industries 15%
reg_df['rnd_dep'] = reg_df['rnd_dep'].apply(lambda x: 0.15 if x==None else x)

## RAW
raw_df['rnd_dep'] = None
for industry in ffi.BEA_rnd_dep:
    raw_df.loc[raw_df['sic_code'] == industry,'rnd_dep'] = ffi.BEA_rnd_dep[industry]
# Rest of industries 15%
raw_df['rnd_dep'] = raw_df['rnd_dep'].apply(lambda x: 0.15 if x==None else x)

## ROBUST_RAW
robust_raw['rnd_dep'] = None
for industry in ffi.BEA_rnd_dep:
    robust_raw.loc[robust_raw['sic_code'] == industry,'rnd_dep'] = ffi.BEA_rnd_dep[industry]
# Rest of industries 15%
robust_raw['rnd_dep'] = robust_raw['rnd_dep'].apply(lambda x: 0.15 if x==None else x)
```

Next, we measure the stock of organization capital by accumulating a fraction of past SG&A spending using the perpetual inventory method. The logic is that at least part of SG&A represents an investment in organization capital through advertising, spending on distribution systems, employee training, and payments to strategy con- sultants. We follow Hulten and Hao (2008), Eisfeldt and Papanikolaou (2014), and Zhang (2014) in counting only 30% of SG&A spending as an investment in intangible capital. We interpret the remaining 70% as operating costs that support the current periodâ€™s profits. We follow Falato, Kadyrzhanova, and Sim (2013) in using a depreciation rate of Î´SG&A = 20%.

```python
## Compute Knowledge Capital
reg_df = fyp.get_know_cap(reg_df)
raw_df = fyp.get_know_cap(raw_df)
robust_raw = fyp.get_know_cap(robust_raw)

## Compute Organization Capital
reg_df = fyp.get_organ_cap(reg_df)
raw_df = fyp.get_organ_cap(raw_df)
robust_raw = fyp.get_organ_cap(robust_raw)
```

```python
# Intangible capital = external I/A (intan) + internal I/A (knowledge cap + organization cap)
reg_df['intan_cap'] = reg_df['intan'] + reg_df['knowledge_cap'] + reg_df['organization_cap']
raw_df['intan_cap'] = raw_df['intan'] + raw_df['knowledge_cap'] + raw_df['organization_cap']
robust_raw['intan_cap'] = robust_raw['intan'] + robust_raw['knowledge_cap'] + robust_raw['organization_cap']

# Intangible capital less goodwill = external I/A (intan) + internal I/A (knowledge cap + organization cap) - GDWL
reg_df['intan_cap_lessgdwl'] = reg_df['intan_cap'] - reg_df['gdwl']
# BV Intangible assets less goodwill = external I/A (intan) - GDWL
reg_df['intan_lessgdwl'] = reg_df['intan'] - reg_df['gdwl']

# Intangible capital less goodwill
robust_raw['intan_cap_lessgdwl'] = robust_raw['intan_cap'] - robust_raw['gdwl']
# BV Intangible assets less goodwill
robust_raw['intan_lessgdwl'] = robust_raw['intan'] - robust_raw['gdwl']
```

# Compute Regression Variables

**Main Paper**: \
Market Capitalization = prcc_f * csho \
Market to Book = Market Capitalization / CEQ \
**Method Paper**: \
We measure the firmâ€™s market value V as the market value of outstanding equity (Compustat items prcc_f times csho), plus the book value of debt (Compustat items dltt + dlc), minus the firmâ€™s current assets (Compustat item act), which include cash, inventory, and marketable securities.

BV_debt = dltt + dlc \
BV_equity = prcc_f * csho \
Market_value_of_firm = BV_debt + BV_equity - act (current asset)

Altman_Z \
A score below 1.8 means it's likely the company is headed for bankruptcy, while companies with scores above 3 are not likely to go bankrupt.
